```python
import requests
from bs4 import BeautifulSoup



# URL of the Telegram channel preview page (https://t.me/s/CHANNLE)
channel_url = 'https://t.me/s/telegram'

# a GET request to fetch the channel page
response = requests.get(channel_url)
response.raise_for_status()  # check we get a successful response

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, 'lxml')

# extract posts
def get_posts(soup):
    posts = []
    
    # Find all the divs containing the posts
    post_divs = soup.find_all('div', class_='tgme_widget_message_bubble')

    # Iterate over posts and extract the post's content
    for post_div in post_divs:
        # Extract text
        post_text = post_div.find('div', class_='tgme_widget_message_text')
        
        # Check the post contains text
        if post_text:
            post_content = post_text.get_text(separator="\n").strip()
            posts.append(post_content)
    
    posts.reverse()
    return posts


posts = get_posts(soup)


for idx, post in enumerate(posts, 1):
    print(f"Post {idx}:\n{post}\n{'-'*40}")
```
